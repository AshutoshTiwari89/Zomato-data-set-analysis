```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
# combining 2 dataset
final_df=pd.merge(d,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
# to check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
#how many records from different -2 countries
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
```

```python
##pie chart-top 3 countries uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
observation-zomato max.records from transaction are from india
then us and uk
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green'])
```

```python
#observation:
1- not rated is very high
2-maximum number of rating are between 2.5 to 3.4
```

```python
## count plot
sns.countplot(x="Rating color",data=ratings, palette=['white','red','orange','yellow','green'])
 
```

```python
## find the countries name that has given 0 ratings
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().head()
```
